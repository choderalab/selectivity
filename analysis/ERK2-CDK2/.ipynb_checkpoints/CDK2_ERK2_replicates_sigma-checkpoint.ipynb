## Make density plot for std vs cc for CDK2/CDK9 for each edge in the map

```python
cdk2_std = []
cdk2_cc = []
cdk2_bar = [] 

for edge in g1.edges:
#     print(edge)
#     print(g1.edges[edge])
#     print(g2.edges[edge])
#     print(g3.edges[edge])
    cdk2_std.append(np.std([g1.edges[edge]['ccc_ddg_val'],g2.edges[edge]['ccc_ddg_val'],g3.edges[edge]['ccc_ddg_val']]))
    cdk2_std.append(np.std([g1.edges[edge]['ccc_ddg_val'],g2.edges[edge]['ccc_ddg_val'],g3.edges[edge]['ccc_ddg_val']]))
    cdk2_std.append(np.std([g1.edges[edge]['ccc_ddg_val'],g2.edges[edge]['ccc_ddg_val'],g3.edges[edge]['ccc_ddg_val']]))
    cdk2_cc.append(g1.edges[edge]['ccc_ddg_unc'])
    cdk2_cc.append(g2.edges[edge]['ccc_ddg_unc'])
    cdk2_cc.append(g3.edges[edge]['ccc_ddg_unc'])
    cdk2_bar.append(g1.edges[edge]['bennett_ddg_unc'])
    cdk2_bar.append(g2.edges[edge]['bennett_ddg_unc'])
    cdk2_bar.append(g3.edges[edge]['bennett_ddg_unc'])
zippedList_cdk2 =  list(zip(cdk2_std, cdk2_cc, cdk2_bar))
df_cdk2 = pd.DataFrame(zippedList_cdk2, columns = ['Standard Deviation' , 'Cycle Closure Error', 'Bennett Error']) 
```

```python
reference = '6'
nedges = len(g1.edges)
nligands = len(g1.nodes)
DeltaDeltaG_CycleClosure_ij_target1 = np.zeros([nligands])
dDeltaDeltaG_CycleClosure_ij_target1 = np.zeros([nligands])
DeltaDeltaG_exp_target1 = np.zeros([nligands])
for edge_index, node in enumerate(g1.nodes):
    DeltaDeltaG_exp_target1[edge_index] = g1.nodes[node]['exp_dg'] - g1.nodes[reference]['exp_dg']
    DeltaDeltaG_CycleClosure_ij_target1[edge_index] = g7.nodes[node]['pred_dg_val'] - g7.nodes[reference]['pred_dg_val']
    dDeltaDeltaG_CycleClosure_ij_target1[edge_index] = g7.nodes[node]['pred_dg_unc']

DeltaDeltaG_CycleClosure_ij_target2 = np.zeros([nligands])
dDeltaDeltaG_CycleClosure_ij_target2 = np.zeros([nligands])
DeltaDeltaG_exp_target2 = np.zeros([nligands])
for edge_index, node in enumerate(g2.nodes):
    DeltaDeltaG_exp_target2[edge_index] = g2.nodes[node]['exp_dg'] - g2.nodes[reference]['exp_dg']
    DeltaDeltaG_CycleClosure_ij_target2[edge_index] = g8.nodes[node]['pred_dg_val'] - g8.nodes[reference]['pred_dg_val']
    dDeltaDeltaG_CycleClosure_ij_target2[edge_index] = g8.nodes[node]['pred_dg_unc']

    
DeltaDeltaG_CycleClosure_ij_target3 = np.zeros([nligands])
dDeltaDeltaG_CycleClosure_ij_target3 = np.zeros([nligands])
DeltaDeltaG_exp_target3 = np.zeros([nligands])
for edge_index, node in enumerate(g3.nodes):
    DeltaDeltaG_exp_target3[edge_index] = g3.nodes[node]['exp_dg'] - g3.nodes[reference]['exp_dg']
    DeltaDeltaG_CycleClosure_ij_target3[edge_index] = g9.nodes[node]['pred_dg_val'] - g9.nodes[reference]['pred_dg_val']
    dDeltaDeltaG_CycleClosure_ij_target3[edge_index] = g9.nodes[node]['pred_dg_unc']
    
DeltaG_exp_CDK2 = np.concatenate((DeltaDeltaG_exp_target3, DeltaDeltaG_exp_target2, DeltaDeltaG_exp_target1), axis=None)
DeltaG_calc_CDK2 = np.concatenate((DeltaDeltaG_CycleClosure_ij_target3, DeltaDeltaG_CycleClosure_ij_target2, DeltaDeltaG_CycleClosure_ij_target1), axis=None)
dDeltaG_calc_CDK2 = np.concatenate((dDeltaDeltaG_CycleClosure_ij_target3, dDeltaDeltaG_CycleClosure_ij_target2, dDeltaDeltaG_CycleClosure_ij_target1), axis=None)
```

```python
nedges = len(g1.edges)
nligands = len(g1.nodes)
DeltaDeltaG_CycleClosure_ij_target1 = np.zeros([nligands])
dDeltaDeltaG_CycleClosure_ij_target1 = np.zeros([nligands])
DeltaDeltaG_exp_target1 = np.zeros([nligands])
for edge_index, node in enumerate(g4.nodes):
    DeltaDeltaG_exp_target1[edge_index] = g4.nodes[node]['exp_dg'] - g4.nodes[reference]['exp_dg']
    DeltaDeltaG_CycleClosure_ij_target1[edge_index] = g10.nodes[node]['pred_dg_val'] - g10.nodes[reference]['pred_dg_val']
    dDeltaDeltaG_CycleClosure_ij_target1[edge_index] = g10.nodes[node]['pred_dg_unc']

DeltaDeltaG_CycleClosure_ij_target2 = np.zeros([nligands])
dDeltaDeltaG_CycleClosure_ij_target2 = np.zeros([nligands])
DeltaDeltaG_exp_target2 = np.zeros([nligands])
for edge_index, node in enumerate(g5.nodes):
    DeltaDeltaG_exp_target2[edge_index] = g5.nodes[node]['exp_dg'] - g5.nodes[reference]['exp_dg']
    DeltaDeltaG_CycleClosure_ij_target2[edge_index] = g11.nodes[node]['pred_dg_val'] - g11.nodes[reference]['pred_dg_val']
    dDeltaDeltaG_CycleClosure_ij_target2[edge_index] = g11.nodes[node]['pred_dg_unc']
    
DeltaDeltaG_CycleClosure_ij_target3 = np.zeros([nligands])
dDeltaDeltaG_CycleClosure_ij_target3 = np.zeros([nligands])
DeltaDeltaG_exp_target3 = np.zeros([nligands])
for edge_index, node in enumerate(g6.node):
    DeltaDeltaG_exp_target3[edge_index] = g6.nodes[node]['exp_dg'] - g6.nodes[reference]['exp_dg']
    DeltaDeltaG_CycleClosure_ij_target3[edge_index] = g12.nodes[node]['pred_dg_val'] - g12.nodes[reference]['pred_dg_val']
    dDeltaDeltaG_CycleClosure_ij_target3[edge_index] = g12.nodes[node]['pred_dg_unc']
    
DeltaG_exp_ERK2 = np.concatenate((DeltaDeltaG_exp_target3, DeltaDeltaG_exp_target2, DeltaDeltaG_exp_target1), axis=None)
DeltaG_calc_ERK2 = np.concatenate((DeltaDeltaG_CycleClosure_ij_target3, DeltaDeltaG_CycleClosure_ij_target2, DeltaDeltaG_CycleClosure_ij_target1), axis=None)
dDeltaG_calc_ERK2 = np.concatenate((dDeltaDeltaG_CycleClosure_ij_target3, dDeltaDeltaG_CycleClosure_ij_target2, dDeltaDeltaG_CycleClosure_ij_target1), axis=None)

labels = []
for node in g1.nodes:
    labels.append(str(node))
for node in g2.nodes:
    labels.append(str(node))
    
for node in g3.nodes:
    labels.append(str(node))
```
