```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context('talk')
sns.set_style('white')
from matplotlib.offsetbox import AnchoredText
```

```python
steps = np.linspace(0,1,50)
sigma = 0.5
noises = [0, .1, .2,.4]
nnoise = len(noises)
nsteps = len(steps)
threshold = 2.8
speedup_data = np.zeros([nnoise, nsteps])

for i,noise in enumerate(noises):
    for j,rho in enumerate(steps):
        sigma_forcefield = np.sqrt(sigma **2 + sigma ** 2 - 2*rho*sigma*sigma)
        sigma_stat = np.sqrt(noise **2 + noise ** 2)
        true = np.random.normal(0,1,40000000)
        forcefield_error = np.random.normal(0, sigma_forcefield, true.shape)
        statistical_error = np.random.normal(0, sigma_stat, true.shape)
        screened = true + forcefield_error + statistical_error
        baseline = len(np.where(true > threshold)[0]) / len(true)
        speedup = (len(np.where(true[np.where(screened > threshold)[0]] > threshold)[0]) / len(true[np.where(screened > threshold)[0]])) / baseline
        speedup_data[i,j] = speedup
np.save('speedup_data_%s.npy' % threshold, speedup_data)
```

```python
speedup_data = np.load('./speedup_data_2.8.npy')
steps = np.linspace(0,1,50)
sigma = 1.0 
noises = [0, .1, .2, .4]
nnoise = len(noises)
nsteps = len(steps)
threshold = 2.8
fig = plt.figure(figsize=[9,6])
ax = plt.subplot(1,1,1)
cmap = sns.cubehelix_palette(5, start=.5, rot=-.75)
for i,noise in enumerate(noises):
    plt.plot(steps, speedup_data[i], label='$\sigma$ = %s' % noise, c=cmap[i], lw=4)
    
plt.legend()
text = AnchoredText('100x threshold', loc=9, pad=.3, prop=dict(fontsize=16))
ax.add_artist(text)
plt.xlim([0,1])
#plt.ylim([0,16])
plt.xticks([0,.5,1], fontsize=15)
#plt.yticks([0,4,8,12,16], fontsize=15)
plt.ylabel('Speedup', fontsize=20)
plt.xlabel('$\\rho$', fontsize=25)
plt.title('Effect of $\\rho$ on speedup', fontsize=22)
fig.savefig('rhovsspeedup_stat.pdf', dpi=400)
```
